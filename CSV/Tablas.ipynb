```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
productos = pd.read_csv("https://github.com/sh4rkd/REPORTE-01-MIRAMONTES-FRED/raw/main/CSV/lifestore_products.csv")
ventas = pd.read_csv("https://github.com/sh4rkd/REPORTE-01-MIRAMONTES-FRED/raw/main/CSV/lifestore_sales.csv")
busquedas = pd.read_csv("https://github.com/sh4rkd/REPORTE-01-MIRAMONTES-FRED/raw/main/CSV/lifestore_searches.csv")
```

```python
total_ventas = ventas.groupby('id_product').size()
total_ventas = pd.DataFrame(total_ventas)
total_ventas = pd.merge(total_ventas,productos,on="id_product")
total_ventas = total_ventas.rename(columns={0:'total_ventas'})
```

```python
busquedas = busquedas.rename(columns={"id product":'id_product'})
busquedas = busquedas.groupby('id_product').size()
busquedas = pd.DataFrame(busquedas)
busquedas = busquedas.rename(columns={0:'total_busquedas'})
```

```python
total_ventas = pd.merge(total_ventas,busquedas,on="id_product")
```

```python
total_ventas.sort_values(by=['total_ventas'],ascending=False)
```

```python
total_ventas.sort_values(by=['total_ventas'],ascending=True)
```

```python
resenias = ventas.groupby('id_product').agg({'score':min,'refund':np.sum})
total_ventas = pd.merge(total_ventas,resenias,on="id_product")
```

```python
total_ventas.sort_values(by=['score'],ascending=False)
```

```python
total_ventas.sort_values(by=['score'],ascending=True)
```

```python
ventas_mensuales = ventas.join(productos.set_index('id_product'), on='id_product')          
ventas_mensuales['date'] = pd.to_datetime(ventas_mensuales['date'])
top_monthly_sales = ventas_mensuales.groupby(ventas_mensuales['date'].dt.strftime('%m')).agg({'price':np.sum}).nlargest(12,columns='price')
top_monthly_sales.columns=['venta_mensual']
```

```python
top_monthly_sales.index = pd.Series(['Marzo', 'Abril', 'Febrero','Enero','Mayo','Junio','Julio','Septiembre','Noviembre','Octubre','Diciembre','Agosto'])
```
